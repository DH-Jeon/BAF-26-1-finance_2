```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("../data/Crawling_Final")

CATEGORIES = ["supply", "demand", "geopolitical", "financial"]

MERGED_FILES = {
    cat: DATA_DIR / f"merged_Oil_{cat.capitalize()}_Final.csv"
    for cat in CATEGORIES
}

category_news_dfs = {}

for cat, file_path in MERGED_FILES.items():
    if not file_path.exists():
        print(f"[{cat}] 파일 없음 → {file_path}")
        continue

    df = pd.read_csv(file_path)

    # 날짜 형식 정리
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

    # headline 결측 제거
    df = df[df["headline"].notna()].copy()

    # category 컬럼 보정
    df["category"] = cat

    category_news_dfs[cat] = df

    print(f"[{cat}] 로드 완료: {len(df):,}건")
    print(f"  기간: {df['date'].min().date()} ~ {df['date'].max().date()}")
```

```python
for cat, df in category_news_dfs.items():
    print(f"\n=== [{cat}] ===")
    display(df.head())
    print(df.columns.tolist())
```

## 1. oil_relevance 필터링

```python
RELEVANCE_THRESHOLD = 1

filtered_category_news_dfs = {}

for cat, df in category_news_dfs.items():
    before = len(df)

    df = df.copy()

    # oil_relevance 숫자형 보장
    # 크롤링 CSV 재로드 시 문자열/헤더 행이 섞이면 coerce로 NaN 처리
    df["oil_relevance"] = pd.to_numeric(df["oil_relevance"], errors="coerce")

    # oil_relevance 기준으로 원유 관련성이 있는 기사만 유지
    df_filtered = df[
        df["oil_relevance"] >= RELEVANCE_THRESHOLD
    ].copy()

    after = len(df_filtered)

    filtered_category_news_dfs[cat] = df_filtered

    print(f"[{cat}] relevance filtering")
    print(f"  before: {before:,}건")
    print(f"  after : {after:,}건")
    print(f"  removed: {before - after:,}건 ({(before - after) / before * 100:.1f}%)")
```

```python
filter_summary = []

for cat in category_news_dfs:
    before = len(category_news_dfs[cat])
    after = len(filtered_category_news_dfs[cat])

    filter_summary.append({
        "category": cat,
        "before": before,
        "after": after,
        "removed": before - after,
        "remaining_rate": after / before if before > 0 else 0,
        "removed_rate": (before - after) / before if before > 0 else 0,
    })

filter_summary_df = pd.DataFrame(filter_summary)

display(
    filter_summary_df.assign(
        remaining_rate=lambda x: (x["remaining_rate"] * 100).round(1),
        removed_rate=lambda x: (x["removed_rate"] * 100).round(1),
    )
)
```

#### 2-2. FinBERT 모델 로드 및 함수 정의

```python
# import sys
# !{sys.executable} -m pip install torch --index-url https://download.pytorch.org/whl/cpu
# !{sys.executable} -m pip install transformers tqdm
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.nn.functional import softmax
from tqdm.auto import tqdm
```

```python
MODEL_NAME = "ProsusAI/finbert"

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"device: {device}")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME).to(device)
model.eval()

id2label = model.config.id2label
print(id2label)
```

```python
def run_finbert(texts, batch_size=32):
    results = []

    for i in tqdm(range(0, len(texts), batch_size)):
        batch_texts = texts[i:i + batch_size]

        inputs = tokenizer(
            batch_texts,
            padding=True,
            truncation=True,
            max_length=128,
            return_tensors="pt"
        ).to(device)

        with torch.no_grad():
            outputs = model(**inputs)
            probs = softmax(outputs.logits, dim=1).cpu().numpy()

        for prob in probs:
            label_probs = {
                id2label[idx].lower(): float(prob[idx])
                for idx in range(len(prob))
            }

            positive_prob = label_probs.get("positive", 0.0)
            neutral_prob = label_probs.get("neutral", 0.0)
            negative_prob = label_probs.get("negative", 0.0)

            sentiment_label = max(label_probs, key=label_probs.get)
            sentiment_score = positive_prob - negative_prob

            results.append({
                "positive_prob": positive_prob,
                "neutral_prob": neutral_prob,
                "negative_prob": negative_prob,
                "sentiment_label": sentiment_label,
                "sentiment_score": sentiment_score,
            })

    return pd.DataFrame(results)
```

```python
### 테스트용
test_texts = finbert_input_dfs["supply"]["finbert_text"].head(5).tolist()
test_result = run_finbert(test_texts, batch_size=5)
display(test_result)
```

#### 2-3. 카테고리별 FinBERT 실행 및 저장

```python
SENTIMENT_DIR = Path("../data/Sentiment_raw")
SENTIMENT_DIR.mkdir(parents=True, exist_ok=True)

finbert_result_dfs = {}

for cat, df in finbert_input_dfs.items():
    print(f"\n=== [{cat}] FinBERT 감성분석 시작: {len(df):,}건 ===")

    texts = df["finbert_text"].tolist()
    sentiment_df = run_finbert(texts, batch_size=32)

    result_df = pd.concat(
        [df.reset_index(drop=True), sentiment_df],
        axis=1
    )

    save_path = SENTIMENT_DIR / f"oil_{cat}_with_finbert.csv"
    result_df.to_csv(save_path, index=False, encoding="utf-8-sig")

    finbert_result_dfs[cat] = result_df

    print(f"[{cat}] 저장 완료 → {save_path}")
```

```python
### FinBERT 결과 확인

for cat, df in finbert_result_dfs.items():
    print(f"\n=== [{cat}] sentiment label 분포 ===")
    print(df["sentiment_label"].value_counts(normalize=True).round(3))
    display(df[[
        "date", "category", "keyword", "headline",
        "positive_prob", "neutral_prob", "negative_prob",
        "sentiment_label", "sentiment_score"
    ]].head())
```

## 3. 카테고리별 일별 aggregation

```python
DAILY_DIR = Path("../data/Sentiment_Final")
DAILY_DIR.mkdir(parents=True, exist_ok=True)

daily_sentiment_dfs = {}

for cat, df in finbert_result_dfs.items():
    df_daily = df.copy()

    # 날짜 형식 정리
    df_daily["date"] = pd.to_datetime(df_daily["date"], errors="coerce")
    df_daily = df_daily[df_daily["date"].notna()].copy()

    # 날짜 단위로 변환
    df_daily["date"] = df_daily["date"].dt.date

    # 일별 aggregation
    # oil_relevance는 진단/검토용 평균값으로 저장하며,
    # sentiment 평균 계산의 가중치로는 사용하지 않음
    daily = (
        df_daily.groupby("date")
        .agg(
            **{
                f"sentiment_{cat}": ("sentiment_score", "mean"),
                f"oil_relevance_{cat}": ("oil_relevance", "mean"),
                f"news_count_{cat}": ("headline", "count"),
                f"positive_prob_{cat}": ("positive_prob", "mean"),
                f"neutral_prob_{cat}": ("neutral_prob", "mean"),
                f"negative_prob_{cat}": ("negative_prob", "mean"),
            }
        )
        .reset_index()
    )

    daily["date"] = pd.to_datetime(daily["date"])

    save_path = DAILY_DIR / f"daily_sentiment_{cat}.csv"
    daily.to_csv(save_path, index=False, encoding="utf-8-sig")

    daily_sentiment_dfs[cat] = daily

    print(f"[{cat}] daily sentiment 저장 완료 → {save_path}")
    print(f"  기간: {daily['date'].min().date()} ~ {daily['date'].max().date()}")
    print(f"  날짜 수: {len(daily):,}")
```

[가중치]

- 키워드 빈도수에 대한 가중치

=> 날짜별로 sentiment_score를 기사 단위 관측치를 기준으로 평균하므로, 특정 키워드에서 더 많은 기사가 수집된 경우 해당 키워드의 sentiment가 일별 지표에 더 크게 반영된다. 따라서 키워드 빈도 기반 가중 효과가 내재되어 있다. 

- Oil_relevance에 대한 가중치

=> 예를 들어 oil_relevance=3이 oil_relevance=1보다 정확히 3배 더 원유시장 관련성이 높다고 보기는 어려운데, 가중평균에 넣으면 실제로는 3배 가중치를 주는 효과가 생겨 가중치를 적용하지 않았다. 

```python
### 일별 뉴스 커버리지 확인

for cat, df in filtered_category_news_dfs.items():
    daily_count = (
        df.assign(date=pd.to_datetime(df["date"]).dt.date)
        .groupby("date")
        .size()
    )

    start_date = pd.to_datetime(df["date"]).min()
    end_date = pd.to_datetime(df["date"]).max()
    total_days = (end_date - start_date).days + 1

    print(f"\n=== [{cat}] ===")
    print(f"뉴스 있는 날짜 수: {len(daily_count):,}")
    print(f"총 기간 달력일 대비 비율: {len(daily_count) / total_days * 100:.1f}%")
    print(f"평균 일별 기사 수: {daily_count.mean():.2f}")
    print(f"최소 일별 기사 수: {daily_count.min()}")
    print(f"최대 일별 기사 수: {daily_count.max()}")
```

## 4. 데이터셋 일 단위 확장

```python
import pandas as pd
import numpy as np
from pathlib import Path

# 경로 설정
SENTIMENT_DIR = Path("../data/Sentiment_Final")

categories = ["demand", "supply", "geopolitical", "financial"]

sentiment_files = {
    cat: SENTIMENT_DIR / f"daily_sentiment_{cat}.csv"
    for cat in categories
}
```

```python
# 공통 날짜표 생성(일별)
calendar_start = min(df["date"].min() for df in daily_sentiment_dfs.values())
calendar_end = max(df["date"].max() for df in daily_sentiment_dfs.values())

calendar_df = pd.DataFrame({
    "date": pd.date_range(start=calendar_start, end=calendar_end, freq="D")
})

print("[Common calendar]")
print(f"start: {calendar_start.date()}")
print(f"end  : {calendar_end.date()}")
print(f"days : {len(calendar_df):,}")
```

```python
# 카테고리별 캘린더 확장
# - news_count는 0으로 채움
# - sentiment는 NaN 유지

calendar_sentiment_dfs = {}

for cat, df in daily_sentiment_dfs.items():
    expanded = calendar_df.merge(df, on="date", how="left")
    
    # 뉴스 없는 날짜 news_count: 0
    expanded[f"news_count_{cat}"] = expanded[f"news_count_{cat}"].fillna(0).astype(int)
    
    # 이진변수(뉴스 없는 날짜(no_news):1, 뉴스 있는 날짜:0)
    expanded[f"no_news_{cat}"] = (expanded[f"news_count_{cat}"] == 0).astype(int)
    
    # sentiment는 NaN 유지
    # oil_relevance, positive_prob 등도 일단 NaN 유지
    
    calendar_sentiment_dfs[cat] = expanded
    
    print(f"[{cat}] calendar expanded")
    print(f"  shape: {expanded.shape}") # 행 수: 6810
    print(f"  date : {expanded['date'].min().date()} ~ {expanded['date'].max().date()}")
    print(f"  no news days: {expanded[f'no_news_{cat}'].sum():,}")
    print()
```

```python
# 확장 결과 확인
expanded_summary = []

for cat, df in calendar_sentiment_dfs.items():
    expanded_summary.append({
        "category": cat,
        "calendar_days": len(df),
        "news_days": (df[f"news_count_{cat}"] > 0).sum(),
        "no_news_days": (df[f"news_count_{cat}"] == 0).sum(),
        "coverage_ratio": (df[f"news_count_{cat}"] > 0).mean(),
        "total_news_count": df[f"news_count_{cat}"].sum(),
        "sentiment_na_count": df[f"sentiment_{cat}"].isna().sum()
    })

expanded_summary = pd.DataFrame(expanded_summary)

display(expanded_summary)
```

## 5. 커버리지 확인

#### 5-1. 카테고리별 뉴스 빈도수

```python
category_news_count = []

for cat, df in calendar_sentiment_dfs.items():
    news_col = f"news_count_{cat}"
    
    category_news_count.append({
        "category": cat,
        "news_days": (df[news_col] > 0).sum(),
        "total_news_count": df[news_col].sum(),
        "avg_news_count_news_days": df.loc[df[news_col] > 0, news_col].mean(),
        "min_news_count_news_days": df.loc[df[news_col] > 0, news_col].min(),
        "max_news_count_news_days": df.loc[df[news_col] > 0, news_col].max()
    })

category_news_count = pd.DataFrame(category_news_count)

display(category_news_count)
```

[해석]

- 카테고리별 뉴스 빈도와 일별 기사 수에 차이가 있어, 감성점수의 안정성이 카테고리별로 다를 수 있다.
- 따라서 단순 일별 감성점수뿐 아니라 7일 평균 감성과 `news_count`를 함께 고려한다.

#### 5-2. 연도별/주별 커버리지

```python
# 연도별 커버리지
yearly_coverage_list = []

for cat, df in calendar_sentiment_dfs.items():
    tmp = df.copy()
    news_col = f"news_count_{cat}"
    
    # 연도 추출
    tmp["year"] = tmp["date"].dt.year
    
    yearly = (
        tmp.groupby("year")
        .agg(
            calendar_days=("date", "count"),
            news_days=(news_col, lambda x: (x > 0).sum()),
            total_news_count=(news_col, "sum"),
            avg_news_count_all_days=(news_col, "mean")
        )
        .reset_index()
    )
    
    yearly["coverage_ratio"] = yearly["news_days"] / yearly["calendar_days"]
    yearly["category"] = cat
    
    yearly_coverage_list.append(yearly)

yearly_coverage = pd.concat(yearly_coverage_list, ignore_index=True)

yearly_coverage = yearly_coverage[
    ["category", "year", "calendar_days", "news_days",
     "coverage_ratio", "total_news_count", "avg_news_count_all_days"]
]

display(yearly_coverage)
```

```python
# 연도별 커버리지 pivot
yearly_coverage_pivot = yearly_coverage.pivot(
    index="year",
    columns="category",
    values="coverage_ratio"
)

display(yearly_coverage_pivot)
```

```python
# 커버리지 0인 연도 확인
zero_coverage_years = yearly_coverage[
    yearly_coverage["coverage_ratio"] == 0
].sort_values(["category", "year"])

display(zero_coverage_years.head(10))

# 커버리지 낮은 연도 확인
low_coverage_threshold = 0.20

low_coverage_years = yearly_coverage[
    (yearly_coverage["coverage_ratio"] > 0) &
    (yearly_coverage["coverage_ratio"] < low_coverage_threshold)
].sort_values(["category", "year"])

display(low_coverage_years.head(10))
```

[해석]

1.  커버리지가 0인 연도: 0
2.  20% 미만의 낮은 커버리지를 보이는 연도: 
- `demand` - 2007~2016년
- `supply` - 2007~2016년
- `financial` - 2007년, 2009년
- `geopolitical` - 0

=> 초기 구간의 감성 피처는 관측 밀도가 낮을 수 있으므로, 이후 7일 평균 생성 및 모델 해석 시 주의가 필요하다.

```python
# 주별 커버리지
weekly_coverage_list = []

for cat, df in calendar_sentiment_dfs.items():
    tmp = df.copy()
    news_col = f"news_count_{cat}"
    
    # 주 시작일 계산(월요일 기준)
    tmp["week_start"] = tmp["date"] - pd.to_timedelta(tmp["date"].dt.weekday, unit="D")
    
    weekly = (
        tmp.groupby("week_start")
        .agg(
            calendar_days=("date", "count"),
            news_days=(news_col, lambda x: (x > 0).sum()), # 해당 주에 뉴스가 있는 날짜 수
            total_news_count=(news_col, "sum"), # 해당 주의 전체 기사 수
            avg_news_count_all_days=(news_col, "mean")
        )
        .reset_index()
    )
    
    weekly["coverage_ratio"] = weekly["news_days"] / weekly["calendar_days"]
    weekly["category"] = cat
    
    weekly_coverage_list.append(weekly)

weekly_coverage = pd.concat(weekly_coverage_list, ignore_index=True)

weekly_coverage = weekly_coverage[
    ["category", "week_start", "calendar_days", "news_days",
     "coverage_ratio", "total_news_count", "avg_news_count_all_days"]
]
```

```python
# 주별 커버리지 pivot
weekly_coverage_pivot = weekly_coverage.pivot(
    index="week_start",
    columns="category",
    values="coverage_ratio"
)

display(weekly_coverage_pivot.head(20))
```

## 6. 평균 감성 피처 생성

#### 6-1. 직전 7일 평균 감성 피처 조합별 테스트


| 항목 | 내용 |
|---|---|
| 기준 기간 | 각 날짜 기준 당일을 제외한 직전 7캘린더일(t-7 ~ t-1) |
| 감성 처리 | 뉴스 없는 날짜의 `sentiment`는 NaN 유지 |
| 평균 방식 | 단순 평균, 기사 수 가중 평균 |
| 비교 기준 | `min_periods = 1, 3, 5` |
| 선택 기준 | 결측 비율과 감성값 안정성 |
| 함께 생성할 피처 | news_{category}_7d, nonews_{category}_7d |

------------------------------------------------------------------------

| 평균 방식 | `min_periods` | 컬럼명 예시 |
|---|---:|---|
| 단순 평균 | 1 | `sent_demand_7d_m1` |
| 단순 평균 | 3 | `sent_demand_7d_m3` |
| 단순 평균 | 5 | `sent_demand_7d_m5` |
| 기사 수 가중 평균 | 1 | `sent_demand_7d_w1` |
| 기사 수 가중 평균 | 3 | `sent_demand_7d_w3` |
| 기사 수 가중 평균 | 5 | `sent_demand_7d_w5` |

```python
# 직전 7일 평균 감성 피처 생성

min_periods_list = [1, 3, 5]
# 1: 직전 7일 중 감성값이 1개라도 있으면 평균 계산
# 3: 감성값이 최소 3개 있어야 평균 계산
# 5: 감성값이 최소 5개 있어야 평균 계산

sentiment_feature_dfs = {}

for cat, df in calendar_sentiment_dfs.items():
    tmp = df.sort_values("date").reset_index(drop=True).copy()
    # 날짜 정렬 후 rolling 적용
    
    sent_col = f"sentiment_{cat}"
    news_col = f"news_count_{cat}"
    
    # 당일 제외
    sent_shift = tmp[sent_col].shift(1)
    news_shift = tmp[news_col].shift(1)
    
    # 직전 7일 기사 수 합계
    tmp[f"news_{cat}_7d"] = (
        news_shift
        .rolling(window=7, min_periods=1)
        .sum()
        .fillna(0)
    )
    
    # 직전 7일 뉴스 없음 여부
    tmp[f"nonews_{cat}_7d"] = (tmp[f"news_{cat}_7d"] == 0).astype(int)

    sent_filled = sent_shift.fillna(0)
    news_eff = news_shift.where(sent_shift.notna(),0)
    
    for mp in min_periods_list:
        # 단순 평균
        tmp[f"sent_{cat}_7d_m{mp}"] = (
            sent_shift
            .rolling(window=7, min_periods=mp)
            .mean()
        )

        # 기사 수 가중 평균
        weighted_sum = (
            (sent_filled * news_eff)
            .rolling(window=7, min_periods=mp)
            .sum()
        )
        
        weight_sum = (
            news_eff
            .rolling(window=7, min_periods=mp)
            .sum()
        )
        
        tmp[f"sent_{cat}_7d_w{mp}"] = (
            weighted_sum / weight_sum.replace(0, np.nan)
            # weight_sum == 0이면 나누기 전에 NaN 처리
        ).replace([np.inf, -np.inf], np.nan)
        # inf, -inf가 혹시 생겨도 최종적으로 NaN 처리
        # 직전 7일 sum(감성점수 × 기사 수) / 직전 7일 sum(기사 수)
    
    sentiment_feature_dfs[cat] = tmp
    
    print(f"[{cat}] 7d sentiment features created")
    print(f"  shape: {tmp.shape}")
    print()
```

```python
# 생성된 피처 확인
for cat, df in sentiment_feature_dfs.items():
    feature_cols = [
        "date",
        f"sentiment_{cat}",
        f"news_count_{cat}", # 원본 일별 기사 수
        f"news_{cat}_7d",
        f"nonews_{cat}_7d",
        f"sent_{cat}_7d_m1",
        f"sent_{cat}_7d_m3",
        f"sent_{cat}_7d_m5",
        f"sent_{cat}_7d_w1",
        f"sent_{cat}_7d_w3",
        f"sent_{cat}_7d_w5",
    ]
    
    print(f"\n=== [{cat}] 7d feature sample ===")
    display(df[feature_cols].head(5))
```

#### 6-2. 7일 평균 감성 조합별 결측치 및 안정성 비교

[확인 기준]

1. 결측치 비율

2. 감성값 안정성

3. 단순 평균과 가중 평균의 차이


```python
sentiment_7d_compare = []

for cat, df in sentiment_feature_dfs.items():
    feature_cols = [
        f"sent_{cat}_7d_m1",
        f"sent_{cat}_7d_m3",
        f"sent_{cat}_7d_m5",

        f"sent_{cat}_7d_w1",
        f"sent_{cat}_7d_w3",
        f"sent_{cat}_7d_w5",
    ]
    
    for col in feature_cols:
        valid = df[col].dropna()
        
        sentiment_7d_compare.append({
            "category": cat,
            "feature": col,
            "missing_ratio": df[col].isna().mean(), # 결측치 비율
            "valid_count": valid.count(), # 결측치가 아닌 값의 개수
            "std": valid.std(), # 결측치를 제외한 감성 피처의 표준편차 -> 안정성 확인
            "q05": valid.quantile(0.05), # 5% 분위수 -> 극단값 확인
            "q95": valid.quantile(0.95), # 95% 분위수 -> 극단값 확인
        })

sentiment_7d_compare = pd.DataFrame(sentiment_7d_compare)

display(sentiment_7d_compare)
```

```python
# 조합별 평균 비교 요약
sentiment_7d_compare["method"] = sentiment_7d_compare["feature"].str.extract(r"7d_([mw]\d)$")

method_summary = (
    sentiment_7d_compare
    .groupby("method")
    .agg(
        avg_missing_ratio=("missing_ratio", "mean"), # 조합별 평균 결측률
        max_missing_ratio=("missing_ratio", "max"), # 조합별 최대 결측률
        avg_std=("std", "mean"), # 조합별 평균 표준편차
        max_std=("std", "max"), # 조합별 최대 표준편차
        avg_valid_count=("valid_count", "mean") # 조합별 평균 유효 관측치 수
    ) 
    .reset_index()
    .sort_values("avg_missing_ratio")
)

display(method_summary)
```

```python
# 단순 평균 vs 가중 평균 상관 분석
mean_weight_corr = []

for cat, df in sentiment_feature_dfs.items():
    for mp in [1, 3, 5]:
        m_col = f"sent_{cat}_7d_m{mp}"
        w_col = f"sent_{cat}_7d_w{mp}"
        
        corr = df[[m_col, w_col]].corr().iloc[0, 1]
        
        mean_weight_corr.append({
            "category": cat,
            "min_periods": mp,
            "mean_feature": m_col,
            "weighted_feature": w_col,
            "corr": corr
        })

mean_weight_corr = pd.DataFrame(mean_weight_corr)

display(mean_weight_corr)
```

[해석]

1. min_periods 비교
- `min_periods=1`: 평균 결측률이 약 13.2%로 가장 낮고, 유효 관측치 수가 가장 많다.
- `min_periods=3`: 평균 결측률이 약 46.0%로 증가하여 데이터 손실이 크다.
- `min_periods=5`: 평균 결측률이 약 65.3%까지 증가하여 최종 피처로 사용하기 어렵다.

=> 최종적으로 결측률과 활용 가능한 관측치 수를 고려하여 `min_periods=1`을 사용한다.

2. 단순 평균 vs 기사 수 가중 평균
- 단순 평균(`m1`): 기사 수를 별도 가중치로 반영하지 않고, 직전 7일 감성값을 단순 평균한 방식
- 기사 수 가중 평균(`w1`): 기사 수를 가중치로 반영한 방식

=> 단순 평균과의 상관계수가 모든 카테고리에서 0.94 이상으로 높게 나타났다.
=> 최종적으로 더 단순하고 해석이 쉬운 단순 평균을 사용한다.

```python
def compute_vif(X: pd.DataFrame) -> pd.Series:
    X_clean = X.dropna()
    if X_clean.shape[0] < X_clean.shape[1] + 5:
        raise ValueError(f"표본 부족: rows={X_clean.shape[0]}, cols={X_clean.shape[1]}")
    Xc = add_constant(X_clean, has_constant="add")
    vifs = pd.Series(
        [variance_inflation_factor(Xc.values, i) for i in range(Xc.shape[1])],
        index=Xc.columns
    )
    return vifs.drop("const").sort_values(ascending=False)

vif_base = compute_vif(df[base_cols])
vif_full = compute_vif(df[base_cols + sentiment_all])

vif_compare = pd.concat(
    [vif_base.rename("VIF_base"), vif_full.rename("VIF_full")],
    axis=1
)
vif_compare["delta"] = vif_compare["VIF_full"] - vif_compare["VIF_base"]
vif_compare = vif_compare.sort_values("VIF_full", ascending=False).round(2)
display(vif_compare)

# 판정: VIF > 10 심각 / > 5 주의
print("\n=== VIF > 10 ===")
display(vif_compare[vif_compare["VIF_full"] > 10])
print("\n=== VIF 5~10 ===")
display(vif_compare[(vif_compare["VIF_full"] > 5) & (vif_compare["VIF_full"] <= 10)])
```

```python
def condition_number(X: pd.DataFrame) -> float:
    Xs = (X - X.mean()) / X.std()  # 표준화 후 계산
    Xs = Xs.dropna()
    return cond(Xs.values)

cn_base = condition_number(df[base_cols])
cn_full = condition_number(df[base_cols + sentiment_all])

print(f"Condition Number (Base)        : {cn_base:>10.1f}")
print(f"Condition Number (Base+Sent)   : {cn_full:>10.1f}")
print(f"증가폭                          : {cn_full - cn_base:>10.1f}")

# 판정: CN > 30 다중공선성 우려 / > 100 심각
```

```python
# 각 sentiment 변수가 base 변수로 얼마나 설명되는가
# R² 높으면 → 해당 sentiment 변수는 base 변수들의 선형결합과 거의 같음 (중복 정보)

import statsmodels.api as sm

partial_r2 = []
for sv in sentiment_all:
    y = df[sv]
    X = df[base_cols + [c for c in sentiment_all if c != sv]]
    valid = y.notna() & X.notna().all(axis=1)
    if valid.sum() < 100:
        continue
    model = sm.OLS(y[valid], add_constant(X[valid])).fit()
    partial_r2.append({"sentiment_var": sv, "R²_by_others": model.rsquared, "VIF_equiv": 1/(1-model.rsquared)})

partial_df = pd.DataFrame(partial_r2).sort_values("R²_by_others", ascending=False)
display(partial_df.round(3))

# R² > 0.9 → 다른 변수로 거의 완전히 설명됨 (중복)
# R² > 0.8 → 정보 중복 상당
```

```python
def diagnose(row):
    vif = row["VIF_full"]
    if vif > 10: return "🔴 심각 (제거/변환 권장)"
    if vif > 5:  return "🟡 주의 (검토)"
    return "🟢 양호"

vif_compare["판정"] = vif_compare.apply(diagnose, axis=1)
display(vif_compare)

print("\n[종합 권고]")
print(f"- Condition Number: {cn_base:.0f} → {cn_full:.0f} ({'악화' if cn_full > cn_base*1.5 else '유지'})")
print(f"- VIF > 10 변수 수: {(vif_compare['VIF_full'] > 10).sum()}")
print(f"- 강한 상관(|r|>0.7) 쌍: {len(corr_pairs)}")
```
